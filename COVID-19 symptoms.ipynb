```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Generación y concatenación DF con todos los síntomas

```python
ns_conj_cong = np.random.binomial (1, 5.0/9, size = 926)
s_conj_cong = np.random.binomial (1, 4.0/9, size = 173)
column_conj_cong = np.concatenate ((ns_conj_cong, s_conj_cong))
```

```python
ns_nasal = np.random.binomial (1, 47.0/53, size = 926)
s_nasal = np.random.binomial (1, 6.0/53, size = 173)
column_nasal = np.concatenate ((ns_nasal, s_nasal))
```

```python
ns_headache = np.random.binomial (1, 124.0/150, size = 926)
s_headache = np.random.binomial (1, 26.0/150, size = 173)
column_headache = np.concatenate ((ns_headache, s_headache))
```

```python
ns_cough = np.random.binomial (1, 623.0/745, size = 926)
s_cough = np.random.binomial (1, 122.0/745, size = 173)
column_cough = np.concatenate ((ns_cough, s_cough))
```

```python
ns_sore = np.random.binomial (1, 130.0/153, size = 926)
s_sore = np.random.binomial (1, 23.0/153, size = 173)
column_sore = np.concatenate ((ns_sore, s_sore))
```

```python
ns_sputum = np.random.binomial (1, 309.0/370, size = 926)
s_sputum = np.random.binomial (1, 61.0/370, size = 173)
column_sputum = np.concatenate ((ns_sputum, s_sputum))
```

```python
ns_fatigue = np.random.binomial (1, 350.0/419, size = 926)
s_fatigue = np.random.binomial (1, 69.0/419, size = 173)
column_fatigue = np.concatenate ((ns_fatigue, s_fatigue))
```

```python
ns_hemoptysis = np.random.binomial (1, 6.0/10, size = 926)
s_hemoptysis = np.random.binomial (1, 4.0/10, size = 173)
column_hemoptysis = np.concatenate ((ns_hemoptysis, s_hemoptysis))
```

```python
ns_short = np.random.binomial (1, 140.0/205, size = 926)
s_short = np.random.binomial (1, 65.0/205, size = 173)
column_short = np.concatenate ((ns_short, s_short))
```

```python
ns_nausea = np.random.binomial (1, 350.0/419, size = 926)
s_nausea = np.random.binomial (1, 350.0/419, size = 173)
column_nausea = np.concatenate ((ns_nausea, s_nausea))
```

```python
ns_diarrhea = np.random.binomial (1, 32.0/42, size = 926)
s_diarrhea = np.random.binomial (1, 10.0/42, size = 173)
column_diarrhea = np.concatenate ((ns_diarrhea, s_diarrhea))
```

```python
ns_myalgia = np.random.binomial (1, 134.0/164, size = 926)
s_myalgia = np.random.binomial (1, 30.0/164, size = 173)
column_myalgia = np.concatenate ((ns_myalgia, s_myalgia))
```

```python
ns_chills = np.random.binomial (1, 100.0/126, size = 926)
s_chills = np.random.binomial (1, 26.0/126, size = 173)
column_chills = np.concatenate ((ns_chills, s_chills))
```

```python
column_total = np.column_stack ((column_conj_cong, column_nasal, column_headache, column_cough, column_sore, column_sputum, column_fatigue, column_hemoptysis, column_short, column_nausea, column_diarrhea, column_myalgia, column_chills))
```

```python
df = pd.DataFrame (column_total)
```

```python
df = df.rename(columns={0: 'Conjunctival congestion', 1 : 'Nasal congestion', 2 : 'Headache', 3 : 'Cough', 4 : 'Sore throat', 5 : 'Sputum production', 6 : 'Fatigue', 7 : 'Hemoptysis', 8 : 'Shortness of breath', 9 : 'Nausea or vomiting', 10 : 'Diarrhea', 11 : 'Myalgia or arthralgia', 12 : 'Chills'})
```

# Histogramas síntomas Severo/No-severo

# Hay correlacion entre todos los sintomas?

```python
symptoms_list= list(df.columns[0:13])

corr = df[symptoms_list].corr() 
plt.figure(figsize=(13,13))
sns.heatmap(corr, cbar = True,  square = True, annot=True, fmt= '.2f',annot_kws={'size': 10},
           xticklabels= symptoms_list, yticklabels= symptoms_list,
           cmap= 'coolwarm')
plt.title ('Correlaciones entre sintomas', size = 20)
plt.show ()
```

```python
# No se observan correlaciones significativas entre síntomas

# Se divide en síntomas respiratorios (Conjunctival congestion, Nasal congestion, Cough, Sore throat, Sputum 
# production, Hemoptysis, Dysnea) 
# y síntomas sístemicos (Headache, Fatigue, Nausea, Diarrhea, Myalgia, Chills)
# Para una visualización más clara de las correlaciones
```

```python
column_respiratory = np.column_stack ((column_conj_cong, column_nasal, column_cough, column_sore, column_sputum, column_hemoptysis, column_short))
df_respiratory = pd.DataFrame (column_respiratory)
df_respiratory = df_respiratory.rename(columns={0: 'Conjunctival congestion', 1 : 'Nasal congestion', 2 : 'Cough', 3 : 'Sore throat', 4 : 'Sputum production', 5 : 'Hemoptysis', 6 : 'Shortness of breath', 7 : 'Nausea or vomiting'})
```

```python
df_respiratory.head ()
```

```python
resp_symptoms_list = list(df_respiratory.columns[:])

corr = df_respiratory[resp_symptoms_list].corr() 
plt.figure(figsize=(10,10))
sns.heatmap(corr, cbar = True,  square = True, annot=True, fmt= '.2f',annot_kws={'size': 15},
           xticklabels= resp_symptoms_list, yticklabels= resp_symptoms_list,
           cmap= 'coolwarm')
plt.title ('Correlaciones entre síntomas respiratorios', size = 20)
plt.show ()
```

```python
column_systemic = np.column_stack ((column_headache, column_fatigue, column_nausea, column_diarrhea, column_myalgia, column_chills))
df_systemic = pd.DataFrame (column_systemic)
df_systemic = df_systemic.rename(columns={0: 'Headache', 2 : 'Fatigue', 3 : 'Nausea or vomiting', 4 : 'Diarrhea', 5 : 'Myalgia or arthralgia', 6 : 'Chills'})

df_systemic.head ()
```

```python
syst_symptoms_list = list(df_systemic.columns[:])

corr = df_systemic[syst_symptoms_list].corr() 
plt.figure(figsize=(10,10))
sns.heatmap(corr, cbar = True,  square = True, annot=True, fmt= '.2f',annot_kws={'size': 15},
           xticklabels= syst_symptoms_list, yticklabels= syst_symptoms_list,
           cmap= 'coolwarm')
plt.title ('Correlaciones entre síntomas "sistémicos"', size = 20)
plt.show ()
```
